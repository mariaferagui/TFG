```python
#https://stackoverflow.com/questions/45065469/using-sklearn-metrics-jaccard-index-with-images
#average : string, [None, ‘binary’ (default), ‘micro’, ‘macro’, ‘samples’, ‘weighted’]

#https://www.programcreek.com/python/?CodeExample=compute+dice
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score
```

```python
## Surface Dice-Sørensen Coefficient
def compute_dice_classes(truth, pred, classes):
    dice_scores = []
    for i in range(classes):
        try:
            pd = np.equal(pred, i)
            gt = np.equal(truth, i)
            dice = 2*np.logical_and(pd, gt).sum()/(pd.sum() + gt.sum())
            dice_scores.append(dice)
        except ZeroDivisionError:
            dice_scores.append(0.0)
    return dice_scores 
```

```python
def compute_dice(truth, pred):
    pd = np.equal(pred, 0)
    gt = np.equal(truth, 0)
    dice_score = 2*np.logical_and(pd, gt).sum()/(pd.sum() + gt.sum())
    return dice_score
```

# Glioblastoma 1

```python
img_true = cv2.imread('glioma_1_tumor.jpg', cv2.IMREAD_GRAYSCALE)
img_pred = cv2.imread('glioma_1_pred_thres.jpg', cv2.IMREAD_GRAYSCALE)

new_size = (img_true.shape[1], img_true.shape[0])
img_pred = cv2.resize(img_pred, new_size)

plt.imshow(img_true); plt.title("Ground tumor"); plt.show()
plt.imshow(img_pred); plt.title("Predicted tumor"); plt.show()
```

```python
img_true_1d = img_true.flatten() 
img_pred_1d = img_pred.flatten()

jaccard_index = jaccard_score(img_true_1d, img_pred_1d, average='micro')
print(jaccard_index)
```

```python
dice = compute_dice(img_true_1d,img_pred_1d)
print (dice)
```
